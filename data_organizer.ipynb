# BEATS.CSV

### IMPORTS

```python
from os import listdir
from os.path import join
import json
```

### DEFAULTS

```python
root = r"C:\Users\david\OneDrive\Documents\DTU\S1\Introduktion til Intelligente Systemer\Eksamensprojekt\BeatData"

LOAD_ALL = True
LOAD_NR = 10
PRINT_NR = 4
```

### LOADING FILES

```python
# List files in folder and limit to LOAD_NR if LOAD_ALL is False
files = listdir(root) if LOAD_ALL else listdir(root)[:LOAD_NR]

# Create empty list for data
data = []

# Loop through files
for file in files:
    with open(join(root, file), 'r') as f:
        file_content = json.loads(f.read())
        song = file.split('-')
        song_year = int(song[0])
        song_id = int(song[1].split('.')[0])

        # Append song_id and song_year to data
        data.append([song_id, song_year])
        
        # Append each number from the list of integers in file_content to the newest list in data
        for number in file_content:
            data[-1].append(number)

# Print the first 10 lines from data
if PRINT_NR is not None:
    for line in data[0:PRINT_NR if PRINT_NR < len(data) else len(data)]:
        print(line)
```

### BEATS.CSV

```python
# Write lines to a csv file
with open('beats.csv', 'w') as f:
    for line in data:
        # If it is the last line, do not add a newline
        if line == data[-1]:
            f.write(';'.join([str(x) for x in line]))
        else:
            f.write(';'.join([str(x) for x in line]) + '\n')
```
